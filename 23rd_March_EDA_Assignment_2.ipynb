```python
#1 Check for non-numeric values
print(df['rating'].unique())

# Convert the 'rating' column to numeric, coercing errors
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Drop rows with NaN ratings (or handle them as needed)
df= df.dropna(subset=['rating'])

# Group by 'category' and calculate the mean of 'rating'
average_ratings = df.groupby('category')['rating'].mean().reset_index()

# Rename columns for clarity
average_ratings.columns = ['category', 'average_rating']

print(average_ratings)
```

```python
#2 What are the top rating_count products by category?
# Ensure 'rating_count' is numeric
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Drop rows where 'rating_count' is NaN
df = df.dropna(subset=['rating_count'])

# Get the index of the top products by rating_count for each category
idx = df.groupby('category')['rating_count'].idxmax()

# Use these indices to get the top products
top_products_by_category = df.loc[idx]

# Display the results
top_products_by_category
```

```python
#3 What is the distribution of discounted prices vs. actual prices?

# Check for missing values in the price columns
df[['discounted_price', 'actual_price']].isnull().sum()

# Plot the distributions
plt.figure(figsize=(14, 6))

# Histogram for discounted prices
plt.subplot(1, 2, 1)
plt.hist(df['discounted_price'].dropna(), bins=30, alpha=0.7, color='blue', label='Discounted Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Discounted Prices')
plt.legend()

# Histogram for actual prices
plt.subplot(1, 2, 2)
plt.hist(df['actual_price'].dropna(), bins=30, alpha=0.7, color='green', label='Actual Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Actual Prices')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#4. How does the average discount percentage vary across categories?

df['discount_percentage'] = df['discount_percentage'].str.rstrip('%').astype(float)

# Group by 'category' and calculate average discount percentage
avg_discount_by_category = df.groupby('category')['discount_percentage'].mean().reset_index()

# Sort by average discount percentage (descending)
avg_discount_by_category = avg_discount_by_category.sort_values(by='discount_percentage', ascending=False)

print(avg_discount_by_category)
```

```python
#5. What are the most popular product names?

# Count the frequency of each product name
popular_products = df['product_name'].value_counts()

# Display the top 10 most popular product names (you can adjust the number as needed)
top_product_names = popular_products.head(10).index.tolist()

print("Top 10 Most Popular Product Names:")
for idx, product_name in enumerate(top_product_names, start=1):
    print(f"{idx}. {product_name}")
```

```python
#6. What are the most popular product keywords?
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
import string

# Assuming 'df' is your DataFrame containing the product data
# Example code snippet for keyword extraction

# Concatenate relevant fields into a single text field
df['text_combined'] = df['product_name'] + ' ' + df['review_title'] + ' ' + df['review_content'] + ' ' + df['about_product']

# Function to clean and tokenize text
def clean_and_tokenize(text):
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Tokenize
    tokens = word_tokenize(text.lower())
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    return tokens

# Apply cleaning and tokenization
df['tokens'] = df['text_combined'].apply(clean_and_tokenize)

# Count word frequencies
word_counts = Counter()
for tokens in df['tokens']:
    word_counts.update(tokens)

# Get most common words (adjust 'n' as needed)
common_words = word_counts.most_common(20)
print(common_words)
```

```python
# Assuming df is your DataFrame with product reviews
# Calculate a weighted score combining rating and rating count
df['weighted_score'] = df['rating'] * df['rating_count']

# Sort reviews by weighted score in descending order to find the most popular ones
popular_reviews = df.sort_values(by='weighted_score', ascending=False)

# You can also consider other factors like engagement metrics, etc.
# For example, if you have 'likes' or 'comments' columns:
# popular_reviews = df.sort_values(by='likes', ascending=False)

# Select top N popular reviews
top_reviews = popular_reviews[['review_id', 'review_title', 'review_content', 'rating', 'rating_count']].head(10)

# Print or return the top reviews
top_reviews
```

```python
#8. What is the correlation between discounted_price and rating?
# Convert 'discounted_price' to numeric type if necessary
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')

# Convert 'rating' to numeric type if necessary
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Calculate the correlation between discounted_price and rating
correlation = df['discounted_price'].corr(df['rating'])

print(f"Correlation between discounted_price and rating: {correlation}")
```

```python
#9. What are the Top 5 categories based on the highest ratings?

# Group by 'category' and calculate the average rating for each category
category_avg_rating = df.groupby('category')['rating'].mean()

# Sort the categories based on average rating in descending order
top_categories = category_avg_rating.sort_values(ascending=False).head(5)

# Print or return the top 5 categories with their average ratings
print("Top 5 categories based on highest ratings:")
top_categories
```

```python
#2.What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram.

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(d['Popularity'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.title('Distribution of Popularity Among Tracks')
plt.grid(True)
plt.show()
```

```python
#3.Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot.

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(d['Duration (ms)'], d['Popularity'], marker='o', color='b', alpha=0.7)

# Add titles and labels
plt.title('Relationship between Popularity and Duration of Tracks')
plt.xlabel('Duration (ms)')
plt.ylabel('Popularity')

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#4.Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot.

# Counting tracks per artist
track_counts = d['Artist'].value_counts()

# Finding the artist with the highest number of tracks
artist_with_most_tracks = track_counts.idxmax()
max_tracks = track_counts.max()

print(f"The artist with the highest number of tracks is '{artist_with_most_tracks}' with {max_tracks} tracks.")

# Plotting the countplot
plt.figure(figsize=(10, 6))
sns.countplot(data=d, x='Artist', order=d['Artist'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Number of Tracks per Artist')
plt.xlabel('Artist')
plt.ylabel('Number of Tracks')
plt.show()
```

```python
#5.What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each.

# Sort the dataset by 'Popularity' in ascending order
df_sorted = d.sort_values(by='Popularity')

# Select the top 5 least popular tracks
top_5_least_popular = df_sorted.head(5)

# Print the Artist and Track Name for each of the top 5 least popular tracks
for index, row in top_5_least_popular.iterrows():
    print(f"Artist: {row['Artist']}, Track Name: {row['Track Name']}")
```

```python
#6.Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist.
# Group by Artist and calculate average popularity
average_popularity = d.groupby('Artist')['Popularity'].mean().reset_index()

# Sort by average popularity in descending order
average_popularity_sorted = average_popularity.sort_values(by='Popularity', ascending=False)

# Get the top 5 most popular artists
top_5_artists = average_popularity_sorted.head(5)

print(top_5_artists)
```

```python
#7.For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist.

# Sort by 'Popularity' in descending order and select top 5
top_artists = d.sort_values(by='Popularity', ascending=False).head(5)

# Iterate through top 5 artists and print their most popular track
for index, row in top_artists.iterrows():
    artist_name = row['Artist']
    most_popular_track = d[d['Artist'] == artist_name].sort_values(by='Popularity', ascending=False).iloc[0]['Track Name']
    print(f"Artist: {artist_name} | Most Popular Track: {most_popular_track}")
```

```python
#9.Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot.

# Calculate average duration per artist
avg_duration_per_artist = d.groupby('Artist')['Duration (ms)'].mean().reset_index()

plt.figure(figsize=(12, 8))
sns.boxplot(x='Artist', y='Duration (ms)', data=d)
plt.xticks(rotation=90)
plt.title('Distribution of Track Durations Across Artists (Box Plot)')
plt.show()


plt.figure(figsize=(12, 8))
sns.violinplot(x='Artist', y='Duration (ms)', data=d)
plt.xticks(rotation=90)
plt.title('Distribution of Track Durations Across Artists (Violin Plot)')
plt.show()
```
