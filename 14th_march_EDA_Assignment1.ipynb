```python
#1. What is the distribution of age among heart failure patients in the dataset?

# Calculate basic statistics
age_stats = data['age'].describe()

# Display statistics
print("Age Distribution Summary:")
print(age_stats)

# Plot a histogram
plt.hist(data['age'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution among Heart Failure Patients')
plt.show()
```

```python
#2.How does the death rate vary with age?

# Grouping data by age and calculating death rate
death_rate_by_age = data.groupby('age')['DEATH_EVENT'].mean().reset_index()

# Plotting the death rate variation with age
plt.figure(figsize=(10, 6))
plt.plot(death_rate_by_age['age'], death_rate_by_age['DEATH_EVENT'], marker='o', linestyle='-')
plt.title('Death Rate Variation with Age')
plt.xlabel('Age')
plt.ylabel('Death Rate')
plt.grid(True)
plt.show()
```

```python
#3 What is the percentage of male and female patients in the dataset?

# Count the number of occurrences of each gender
gender_counts = data['sex'].value_counts()

# Calculate the percentage of each gender
total_patients = len(data)
male_percentage = (gender_counts[1] / total_patients) * 100
female_percentage = (gender_counts[0] / total_patients) * 100

print("Percentage of male patients:", male_percentage)
print("Percentage of female patients:", female_percentage)
```

```python
#4 how does the platelet count vary among different age groups?

# Define age groups
age_groups = pd.cut(data['age'], bins=[0, 20, 40, 60, 80, 100])

# Group data by age groups and calculate platelet count statistics
platelet_stats = data.groupby(age_groups)['platelets'].agg(['mean', 'median', 'std'])

# Visualize the data
plt.figure(figsize=(10, 6))
platelet_stats['mean'].plot(kind='bar', yerr=platelet_stats['std'], capsize=5)
plt.xlabel('Age Groups')
plt.ylabel('Mean Platelet Count')
plt.title('Mean Platelet Count Across Different Age Groups')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#5.Is there a correlation between creatinine and sodium levels in the blood?

creatinine = data['serum_creatinine']
sodium = data['serum_sodium']

# Calculate Pearson correlation coefficient
pearson_corr = creatinine.corr(sodium, method='pearson')

# Calculate Spearman rank correlation coefficient
spearman_corr = creatinine.corr(sodium, method='spearman')

print("Pearson correlation coefficient:", pearson_corr)
print("Spearman rank correlation coefficient:", spearman_corr)
```

```python
#6 how does the prevalence of high blood pressure differ between male and female patients?

# Calculate the prevalence of high blood pressure among male patients
male_high_bp_prevalence = data[data['sex'] == 1]['high_blood_pressure'].mean()

# Calculate the prevalence of high blood pressure among female patients
female_high_bp_prevalence = data[data['sex'] == 0]['high_blood_pressure'].mean()

print("Prevalence of high blood pressure among male patients:", male_high_bp_prevalence)
print("Prevalence of high blood pressure among female patients:", female_high_bp_prevalence)
```

```python
#7 what is the relationship between smoking habits and the occurrence of heart failure?

from scipy.stats import chi2_contingency
import statsmodels.api as sm
import statsmodels.formula.api as smf



# Descriptive Statistics
print(data['smoking'].value_counts())
print(data['DEATH_EVENT'].value_counts())

# Contingency Table
contingency_table = pd.crosstab(data['smoking'], data['DEATH_EVENT'])
print(contingency_table)

# Chi-Square Test
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-Square Test:\nChi2: {chi2}\np-value: {p}\nDegrees of Freedom: {dof}\nExpected Frequencies: \n{expected}")

# Logistic Regression
data['intercept'] = 1  # Add intercept for logistic regression
logit_model = sm.Logit(data['DEATH_EVENT'], data[['smoking', 'intercept']])
result = logit_model.fit()

print(result.summary())

# Logistic Regression with other variables
formula = 'DEATH_EVENT ~ smoking + age + anaemia + creatinine_phosphokinase + diabetes + ejection_fraction + high_blood_pressure + platelets + serum_creatinine + serum_sodium + sex + time'
logit_model_full = smf.logit(formula, data=data).fit()

print(logit_model_full.summary())
```

```python
#8 Are there any noticeable patterns in the distribution of death events across different age groups?

# Bin the age data
bins = [0, 40, 50, 60, 70, 80, 90, 100]
labels = ['<40', '40-49', '50-59', '60-69', '70-79', '80-89', '90+']
data['age_group'] = pd.cut(data['age'], bins=bins, labels=labels, right=False)

# Calculate death event rate for each age group
death_event_rate = data.groupby('age_group')['DEATH_EVENT'].mean()

# Plot the results
death_event_rate.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Age Group')
plt.ylabel('Death Event Rate')
plt.title('Death Event Rate by Age Group')
plt.show()
```

```python
#9 Is there any significant difference in ejection fraction between patients with and without diabetes?

import scipy.stats as stats

# Filter data for patients with and without diabetes
ef_diabetes = data[data['diabetes'] == 1]['ejection_fraction']
ef_no_diabetes = data[data['diabetes'] == 0]['ejection_fraction']

# Perform independent samples t-test
t_statistic, p_value = stats.ttest_ind(ef_diabetes, ef_no_diabetes)

# Print the results
print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("There is a significant difference in ejection fraction between patients with and without diabetes.")
else:
    print("There is no significant difference in ejection fraction between patients with and without diabetes.")
```

```python
#10 how does the serum creatinine level vary between patients who survived and those who did not?


# Select relevant columns for analysis
df_analysis = data[['serum_creatinine', 'DEATH_EVENT']]

# Separate serum creatinine levels for each group
serum_creatinine_survived = df_analysis[df_analysis['DEATH_EVENT'] == 0]['serum_creatinine']
serum_creatinine_not_survived = df_analysis[df_analysis['DEATH_EVENT'] == 1]['serum_creatinine']

# Calculate summary statistics
survived_mean = serum_creatinine_survived.mean()
not_survived_mean = serum_creatinine_not_survived.mean()

# Plot the distribution of serum creatinine levels
plt.figure(figsize=(10, 6))
sns.histplot(serum_creatinine_survived, color='blue', label='Survived', kde=True)
sns.histplot(serum_creatinine_not_survived, color='red', label='Not Survived', kde=True)
plt.axvline(survived_mean, color='blue', linestyle='dashed', linewidth=1.5, label='Mean - Survived')
plt.axvline(not_survived_mean, color='red', linestyle='dashed', linewidth=1.5, label='Mean - Not Survived')
plt.xlabel('Serum Creatinine Level')
plt.ylabel('Count')
plt.title('Distribution of Serum Creatinine Level by Survival Status')
plt.legend()
plt.show()

# Print summary statistics
print(f"Mean serum creatinine - Survived: {survived_mean:.2f}")
print(f"Mean serum creatinine - Not Survived: {not_survived_mean:.2f}")
```
