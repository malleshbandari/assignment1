```python
#1 Define a doubly linked list  [ Will be done in the class ]

#Node Definition
#First, we define the Node class to represent each node in the list.
#Doubly Linked List Definition
#Next, we define the DoublyLinkedList class to manage the linked list operations.

class Node:
    def __init__(self, data):
        self.data = data      # The data stored in the node
        self.next = None      # Pointer to the next node
        self.prev = None      # Pointer to the previous node

class DoublyLinkedList:
    def __init__(self):
        self.head = None  # Pointer to the first node in the list
        self.tail = None  # Pointer to the last node in the list

    def append(self, data):
        new_node = Node(data)  # Create a new node
        if self.head is None:  # If the list is empty
            self.head = new_node
            self.tail = new_node
        else:  # If the list is not empty
            self.tail.next = new_node  # Link the new node to the end of the list
            new_node.prev = self.tail  # Link the end of the list to the new node
            self.tail = new_node       # Update the tail to the new node

    def display_forward(self):
        current = self.head
        while current:  # Traverse the list forward
            print(current.data, end=" ")
            current = current.next
        print()

    def display_backward(self):
        current = self.tail
        while current:  # Traverse the list backward
            print(current.data, end=" ")
            current = current.prev
        print()

        
# Create a doubly linked list
dll = DoublyLinkedList()

# Append data to the list
dll.append(1)
dll.append(2)
dll.append(3)

# Display the list forward
print("Forward:")
dll.display_forward()  # Output: 1 2 3

# Display the list backward
print("Backward:")
dll.display_backward()  # Output: 3 2 1

#Explanation

#Node Class:
#data: Stores the value of the node.
#next: Points to the next node in the list.
#prev: Points to the previous node in the list.

#DoublyLinkedList Class:
#head: Points to the first node in the list.
#tail: Points to the last node in the list.
#append(data): Adds a new node with the given data to the end of the list.
#display_forward(): Prints the values of the nodes from the head to the tail.
#display_backward(): Prints the values of the nodes from the tail to the head.
```

```python
#2 Write a function to reverse a linked list in-place
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def reverse_linked_list(head):
    prev = None
    current = head

    while current is not None:
        next_node = current.next  # store the next node
        current.next = prev       # reverse the link
        prev = current            # move prev to current
        current = next_node       # move to the next node

    return prev  # prev will be the new head


# Helper function to print the list
def print_list(head):
    current = head
    while current is not None:
        print(current.value, end=" -> ")
        current = current.next
    print("None")

# Create a linked list 1 -> 2 -> 3 -> 4 -> None
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4))))

print("Original List:")
print_list(head)

# Reverse the linked list
reversed_head = reverse_linked_list(head)

print("Reversed List:")
print_list(reversed_head)
```

```python
#3.Detect cycle in a linked list.
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def has_cycle(head):
    if not head:
        return False

    slow = head
    fast = head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

        if slow == fast:
            return True

    return False




# Helper function to create a cycle in the linked list
def create_cycle(head, pos):
    if pos == -1:
        return
    cycle_node = head
    for _ in range(pos):
        cycle_node = cycle_node.next
    last_node = head
    while last_node.next:
        last_node = last_node.next
    last_node.next = cycle_node

# Create a linked list 1 -> 2 -> 3 -> 4 -> None
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4))))

# Create a cycle at position 1 (0-based index)
create_cycle(head, 1)

# Detect cycle
print(has_cycle(head))  # Output: True

# Create a linked list without a cycle 1 -> 2 -> 3 -> 4 -> None
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4))))

# Detect cycle
print(has_cycle(head))  # Output: False
```

```python
#Merge two sorted linked list into one
#1->3->5->6->null and 2->4->6->8->null should be merged to make
#1->2->3->4->5->6->7->8

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def mergeTwoLists(l1, l2):
    dummy = ListNode()  # Create a dummy node
    tail = dummy  # Tail pointer for the merged list

    while l1 and l2:
        if l1.val < l2.val:
            tail.next = l1
            l1 = l1.next
        else:
            tail.next = l2
            l2 = l2.next
        tail = tail.next

    if l1:
        tail.next = l1
    if l2:
        tail.next = l2

    return dummy.next  # The merged list is next to the dummy node

# Helper function to print the linked list
def printList(node):
    while node:
        print(node.val, end=" -> ")
        node = node.next
    print("null")

# Creating first sorted linked list: 1 -> 3 -> 5 -> 6 -> null
l1 = ListNode(1)
l1.next = ListNode(3)
l1.next.next = ListNode(5)
l1.next.next.next = ListNode(6)

# Creating second sorted linked list: 2 -> 4 -> 6 -> 8 -> null
l2 = ListNode(2)
l2.next = ListNode(4)
l2.next.next = ListNode(6)
l2.next.next.next = ListNode(8)

# Merging the two sorted linked lists
merged_list = mergeTwoLists(l1, l2)

# Printing the merged linked list
printList(merged_list)
```

```python
#4.Write a function to remove nth node from the end in a linked list
#1->2->3->4->5->6, removing 2nd node from end will return 1->2->3->4->6

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def remove_nth_from_end(head, n):
    dummy = ListNode(0)
    dummy.next = head
    first = second = dummy
    
    # Move first pointer ahead by n nodes
    for _ in range(n + 1):
        first = first.next
    
    # Move both pointers until first reaches the end
    while first is not None:
        first = first.next
        second = second.next
    
    # Remove the nth node from the end
    second.next = second.next.next
    
    return dummy.next

# Helper function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" ")
        current = current.next
    print()

# Example usage:
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)
head.next.next.next.next.next = ListNode(6)

n = 2
head = remove_nth_from_end(head, n)
print_linked_list(head)  # Output: 1 2 3 4 6
```

```python
#5.Remove duplicates from a sorted linked list
#1->2->3->3->4->4->4->5  should be changed to 1->2->3->4->5

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def delete_duplicates(head):
    current = head
    
    while current and current.next:
        if current.val == current.next.val:
            current.next = current.next.next
        else:
            current = current.next
            
    return head

# Helper function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" ")
        current = current.next
    print()

# Example usage:
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(3)
head.next.next.next.next = ListNode(4)
head.next.next.next.next.next = ListNode(4)
head.next.next.next.next.next.next = ListNode(4)
head.next.next.next.next.next.next.next = ListNode(5)

head = delete_duplicates(head)
print_linked_list(head)  # Output: 1 2 3 4 5
```

```python
#Find the intersection of the two linked lists
#1->2->3->4->8->6->9  5->1->6->7  , intersection 1->6

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def getIntersectionNode(headA, headB):
    pointerA = headA
    pointerB = headB

    while pointerA != pointerB:
        pointerA = pointerA.next if pointerA else headB
        pointerB = pointerB.next if pointerB else headA

    return pointerA

# Define linked lists
listA = ListNode(1)
listA.next = ListNode(2)
listA.next.next = ListNode(3)
listA.next.next.next = ListNode(4)
listA.next.next.next.next = ListNode(8)
listA.next.next.next.next.next = ListNode(6)
listA.next.next.next.next.next.next = ListNode(9)

listB = ListNode(5)
listB.next = listA.next.next.next.next  # Connecting listB to the common node

# Find the intersection node
intersection_node = getIntersectionNode(listA, listB)
if intersection_node:
    print("Intersection node value:", intersection_node.val)
else:
    print("No intersection node found.")
```

```python
#Rotate a linked list by k positions to the right
#1->2->3->4->8->6->9 , after rotating for 2 times becomes , 3->4->8->6->9->1->2

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def rotateRight(head, k):
    if not head or k == 0:
        return head
    
    # Calculate the length of the linked list
    length = 1
    current = head
    while current.next:
        length += 1
        current = current.next
    
    # If k is greater than the length, take modulo
    k %= length
    
    # If k becomes zero after modulo, no rotation needed
    if k == 0:
        return head
    
    # Find the (length - k)th node
    fast = head
    for _ in range(length - k - 1):
        fast = fast.next
    
    # Update pointers
    new_head = fast.next
    fast.next = None
    current.next = head
    
    return new_head

# Example usage:
# Construct the linked list
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(8)
head.next.next.next.next.next = ListNode(6)
head.next.next.next.next.next.next = ListNode(9)

# Rotate the list by 2 positions to the right
k = 2
rotated_head = rotateRight(head, k)

# Print the rotated list
current = rotated_head
while current:
    print(current.val, end=" -> ")
    current = current.next
```

```python
#Add Two Numbers Represented by LinkedLists:
#Given two non-empty linked lists representing two non-negative integers, where the digits are stored in reverse order, add the two numCers and return it as a linked list.

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def addTwoNumbers(l1, l2):
    dummy = ListNode()
    current = dummy
    carry = 0
    
    while l1 or l2:
        # If l1 has a node, get its value, otherwise set it to 0
        val1 = l1.val if l1 else 0
        # If l2 has a node, get its value, otherwise set it to 0
        val2 = l2.val if l2 else 0
        
        # Calculate sum and carry
        total = val1 + val2 + carry
        carry = total // 10
        
        # Append the sum % 10 to the result linked list
        current.next = ListNode(total % 10)
        current = current.next
        
        # Move to the next nodes in both linked lists if they exist
        if l1:
            l1 = l1.next
        if l2:
            l2 = l2.next
    
    # If there is still a carry, append it as a new node
    if carry > 0:
        current.next = ListNode(carry)
    
    # Return the next node of the dummy node (the actual head of the result linked list)
    return dummy.next

# Example usage:
# Construct two linked lists representing 243 and 564 (stored in reverse order)
l1 = ListNode(3)
l1.next = ListNode(4)
l1.next.next = ListNode(2)

l2 = ListNode(4)
l2.next = ListNode(6)
l2.next.next = ListNode(5)

# Add the two numbers
result = addTwoNumbers(l1, l2)

# Print the result
while result:
    print(result.val, end=" -> ")
    result = result.next
```

```python
#Clone a Linked List with next and Random Pointer
#Given a linked list of size N where each node has two links: one pointer points to the next node and the
#second pointer points to any node in the list. The task is to create a clone of this linked list in O(N) time. 

#Note: The pointer pointing to the next node is ‘next‘ pointer and the one pointing to an arbitrary node is called ‘arbit’ pointer as it can point to any arbitrary node in the linked list. 

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.random = None

def cloneLinkedList(head):
    if not head:
        return None
    
    # Step 1: Duplicate each node and insert it next to the original node
    current = head
    while current:
        new_node = Node(current.data)
        new_node.next = current.next
        current.next = new_node
        current = new_node.next
    
    # Step 2: Set random pointers for the newly created nodes
    current = head
    while current:
        if current.random:
            current.next.random = current.random.next
        current = current.next.next
    
    # Step 3: Separate the original list from the cloned list
    cloned_head = head.next
    current = head
    while current.next:
        temp = current.next
        current.next = temp.next
        current = temp
    
    return cloned_head

# Example usage:
# Create the original linked list
original_head = Node(1)
original_head.next = Node(2)
original_head.next.next = Node(3)
original_head.next.next.next = Node(4)
original_head.next.next.next.next = Node(5)

# Set random pointers
original_head.random = original_head.next.next  # 1's random points to 3
original_head.next.random = original_head  # 2's random points to 1
original_head.next.next.random = original_head.next.next.next  # 3's random points to 4
original_head.next.next.next.random = original_head.next  # 4's random points to 2
original_head.next.next.next.next.random = original_head 
# Clone the linked list
cloned_head = cloneLinkedList(original_head)

# Verify the cloned list
print("Original List:")
current = original_head
while current:
    random_data = current.random.data if current.random else None
    print(f"Data: {current.data}, Random: {random_data}")
    current = current.next

print("\nCloned List:")
current = cloned_head
while current:
    random_data = current.random.data if current.random else None
    print(f"Data: {current.data}, Random: {random_data}")
    current = current.next
```
